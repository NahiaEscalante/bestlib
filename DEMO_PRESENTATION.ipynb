## üîß Setup

````python
import pandas as pd
import numpy as np
from BESTLIB import ReactiveMatrixLayout
from sklearn.datasets import load_iris, load_wine

np.random.seed(42)
print("‚úÖ BESTLIB cargado correctamente")
````

## üìä Datos

````python
# Dataset Iris (flores)
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target_names[iris.target]
df_iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

print("üå∫ Iris Dataset:")
print(df_iris.head())
print(f"\nüìà Shape: {df_iris.shape}")
print(f"üè∑Ô∏è  Species: {df_iris['species'].unique()}")
````

````python
# Dataset Wine (vinos)
wine = load_wine()
df_wine = pd.DataFrame(wine.data, columns=wine.feature_names)
df_wine['wine_class'] = wine.target_names[wine.target]

print("üç∑ Wine Dataset:")
print(df_wine[['alcohol', 'malic_acid', 'ash', 'wine_class']].head())
print(f"\nüìà Shape: {df_wine.shape}")
print(f"üè∑Ô∏è  Classes: {df_wine['wine_class'].unique()}")
````

---
# üé® DEMOS INTERACTIVOS

## Demo 1: üñ±Ô∏è Brushing B√°sico

**Instrucciones:**
1. **Arrastra el mouse** sobre el scatter plot (arriba)
2. Observa c√≥mo el boxplot (abajo) se actualiza autom√°ticamente
3. Los puntos seleccionados se resaltan en ambos gr√°ficos

````python
demo1 = ReactiveMatrixLayout("""
S
X
""")  # SelectionModel se crea autom√°ticamente

demo1.set_data(df_iris)

demo1.add_scatter('S', 
                  x_col='petal_length', 
                  y_col='petal_width',
                  color_col='species',
                  xLabel='Longitud P√©talo (cm)', 
                  yLabel='Ancho P√©talo (cm)',
                  title='üéØ SCATTER: Arrastra para seleccionar')

demo1.add_boxplot('X', 
                  column='petal_length', 
                  category_col='species',
                  linked_to='S',
                  xLabel='Especies', 
                  yLabel='Longitud P√©talo (cm)',
                  title='üì¶ BOXPLOT: Actualizado din√°micamente')

demo1.display()
print("\n‚úÖ Demo 1 listo. ¬°Prueba el brushing!")
````

---
## Demo 2: üîÑ Interacci√≥n Scatter ‚Üî Bar

**Instrucciones:**
1. **Arrastra el mouse** sobre el scatter plot (arriba) para seleccionar puntos
2. Observa c√≥mo el bar chart (abajo) se **actualiza autom√°ticamente** mostrando solo las especies seleccionadas
3. Los datos se re-agregan din√°micamente seg√∫n tu selecci√≥n

> üí° **Nota**: Esta demo muestra la re-agregaci√≥n autom√°tica de datos. Cuando seleccionas puntos en el scatter, el sistema cuenta cu√°ntos puntos de cada especie seleccionaste y actualiza las barras en tiempo real.

````python
demo2 = ReactiveMatrixLayout("""
S
B
""")  # SelectionModel se crea autom√°ticamente

demo2.set_data(df_iris)

# Scatter plot como vista principal (interactivo con brush)
demo2.add_scatter('S', 
                  x_col='sepal_length', 
                  y_col='sepal_width',
                  color_col='species',
                  interactive=True,
                  xLabel='Longitud S√©palo', 
                  yLabel='Ancho S√©palo',
                  title='üéØ SCATTER: Selecciona puntos (arrastra)')

# Bar chart enlazado al scatter (se actualiza autom√°ticamente)
demo2.add_barchart('B', 
                   category_col='species', 
                   value_col='petal_length',
                   linked_to='S',
                   xLabel='Especies', 
                   yLabel='Longitud promedio',
                   title='üìä BAR CHART: Actualizado por selecci√≥n')

demo2.display()
print("\n‚úÖ Demo 2 listo. ¬°Selecciona puntos en el scatter!")
````

---
## Demo 3: üéõÔ∏è Dashboard Completo (5 vistas enlazadas)

**Instrucciones:**
1. **Selecciona puntos** en el scatter plot grande (izquierda)
2. Observa c√≥mo **TODAS** las dem√°s vistas se actualizan
3. Cada gr√°fico muestra diferentes aspectos de los datos seleccionados

````python
dashboard = ReactiveMatrixLayout("""
S S B
H X V
""")  # SelectionModel se crea autom√°ticamente

dashboard.set_data(df_iris)

# Scatter principal (control)
dashboard.add_scatter('S', 
                      x_col='petal_length', 
                      y_col='petal_width',
                      color_col='species',
                      xLabel='Longitud P√©talo', 
                      yLabel='Ancho P√©talo',
                      title='üéØ CONTROL: Selecciona aqu√≠')

# Bar Chart enlazado
dashboard.add_barchart('B', 
                       category_col='species', 
                       value_col='petal_length',
                       linked_to='S',
                       xLabel='Especies',
                       title='üìä Conteo')

# Histogram enlazado
dashboard.add_histogram('H', 
                        column='sepal_length', 
                        bins=15,
                        linked_to='S',
                        xLabel='Longitud S√©palo',
                        title='üìà Distribuci√≥n')

# Boxplot enlazado
dashboard.add_boxplot('X', 
                      column='sepal_width', 
                      category_col='species',
                      linked_to='S',
                      xLabel='Especies',
                      title='üì¶ Estad√≠sticas')

# Violin plot enlazado
dashboard.add_violin('V', 
                     value_col='petal_length', 
                     category_col='species',
                     linked_to='S',
                     xLabel='Especies',
                     title='üéª Densidad')

dashboard.display()
print("\n‚úÖ Dashboard completo. ¬°5 vistas sincronizadas!")
````

---
## Demo 4: üîó M√∫ltiples Scatter Plots Enlazados

**Instrucciones:**
1. Selecciona en el scatter de la **izquierda**
2. El scatter de la **derecha** muestra los mismos puntos
3. √ötil para explorar correlaciones entre diferentes variables

````python
demo4 = ReactiveMatrixLayout("""
S1 S2
""")  # SelectionModel se crea autom√°ticamente

demo4.set_data(df_iris)

demo4.add_scatter('S1', 
                  x_col='sepal_length', 
                  y_col='sepal_width',
                  color_col='species',
                  xLabel='Longitud S√©palo',
                  yLabel='Ancho S√©palo',
                  title='üîç Vista 1: S√©palo')

demo4.add_scatter('S2', 
                  x_col='petal_length', 
                  y_col='petal_width',
                  color_col='species',
                  linked_to='S1',
                  xLabel='Longitud P√©talo',
                  yLabel='Ancho P√©talo',
                  title='üìç Vista 2: P√©talo')

demo4.display()
print("\n‚úÖ Demo 4 listo. Explora las correlaciones!")
````

---
## Demo 5: üç∑ Dashboard Avanzado - Wine Dataset

**Dataset diferente con 6 vistas enlazadas**

**Instrucciones:**
1. Selecciona regiones en el scatter principal
2. Todas las vistas se actualizan mostrando caracter√≠sticas del vino seleccionado
3. Observa patrones en alcohol, acidez m√°lica, y otras propiedades

````python
wine_dashboard = ReactiveMatrixLayout("""
S S H
B X P
""")  # SelectionModel se crea autom√°ticamente

wine_dashboard.set_data(df_wine)

# Scatter principal
wine_dashboard.add_scatter('S', 
                           x_col='alcohol', 
                           y_col='malic_acid',
                           color_col='wine_class',
                           xLabel='Alcohol (%)',
                           yLabel='√Åcido M√°lico',
                           title='üç∑ Alcohol vs Acidez M√°lica')

# Histogram de alcohol
wine_dashboard.add_histogram('H', 
                             column='alcohol', 
                             bins=20,
                             linked_to='S',
                             xLabel='Alcohol (%)',
                             title='üìä Distribuci√≥n Alcohol')

# Bar chart de clases
wine_dashboard.add_barchart('B', 
                            category_col='wine_class', 
                            value_col='alcohol',
                            linked_to='S',
                            xLabel='Clase de Vino',
                            title='üìä Alcohol por Clase')

# Boxplot de acidez
wine_dashboard.add_boxplot('X', 
                           column='malic_acid', 
                           category_col='wine_class',
                           linked_to='S',
                           xLabel='Clase',
                           title='üì¶ Acidez M√°lica')

# Pie chart de proporci√≥n
wine_dashboard.add_pie('P', 
                       category_col='wine_class', 
                       value_col='alcohol',
                       linked_to='S',
                       title='ü•ß Proporci√≥n de Clases')

wine_dashboard.display()
print("\n‚úÖ Wine Dashboard listo. ¬°6 vistas sincronizadas!")
````

## üß™ BONUS: KDE Test

Prueba del gr√°fico KDE (Kernel Density Estimation):

````python
# Test SIMPLE del KDE - Ejemplo funcional
import pandas as pd
from sklearn.datasets import load_iris

# Datos simples
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

print("üìä Creando KDE de petal_length...")
print(f"   Datos: {len(df_iris)} filas")
print(f"   Rango: {df_iris['petal_length'].min():.2f} - {df_iris['petal_length'].max():.2f}")

# Crear gr√°fico KDE
from BESTLIB import MatrixLayout

kde_test = MatrixLayout("K")
kde_test.map_kde('K', df_iris, column='petal_length', color='#e74c3c', fill=True)
kde_test.display()

print("\n‚úÖ Si ves una curva roja, ¬°KDE funciona!")
````
