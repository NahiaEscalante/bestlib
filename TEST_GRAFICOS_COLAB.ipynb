## Imports y Setup

```python
import pandas as pd
import numpy as np
from BESTLIB import MatrixLayout, ReactiveMatrixLayout, SelectionModel
from sklearn.datasets import load_iris, load_wine

np.random.seed(42)
```

## Datos de Prueba

```python
# Dataset Iris
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target_names[iris.target]
df_iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

print("Dataset Iris:")
print(df_iris.head())
print(f"Shape: {df_iris.shape}")
```

```python
# Dataset Wine
wine = load_wine()
df_wine = pd.DataFrame(wine.data, columns=wine.feature_names)
df_wine['wine_class'] = wine.target_names[wine.target]

print("Dataset Wine:")
print(df_wine[['alcohol', 'malic_acid', 'ash', 'wine_class']].head())
print(f"Shape: {df_wine.shape}")
```

```python
# Datasets sint√©ticos
n_points = 100
df_time = pd.DataFrame({
    'time': pd.date_range('2024-01-01', periods=n_points, freq='D'),
    'value': np.cumsum(np.random.randn(n_points)),
    'value2': np.cumsum(np.random.randn(n_points) * 0.8),
    'category': np.random.choice(['A', 'B', 'C'], n_points)
})
df_time['time_numeric'] = range(n_points)

df_error = pd.DataFrame({
    'x': range(1, 11),
    'y': [2, 3, 5, 4, 6, 8, 7, 9, 11, 10],
    'error': [0.5, 0.3, 0.6, 0.4, 0.5, 0.7, 0.6, 0.8, 0.9, 0.7]
})

print("Datasets sint√©ticos creados ‚úì")
```

# PARTE 1: GR√ÅFICOS INDIVIDUALES (26 tipos)

## 1. Scatter Plot

```python
layout1 = MatrixLayout("S")
layout1.map_scatter('S', df_iris, x_col='petal_length', y_col='petal_width',
                    color_col='species', xLabel='Petal Length', yLabel='Petal Width',
                    title='Scatter Plot: Iris Dataset')
layout1.display()
```

## 2. Bar Chart

```python
df_bar = df_iris.groupby('species')['petal_length'].mean().reset_index()
df_bar.columns = ['species', 'avg_petal_length']

layout2 = MatrixLayout("B")
layout2.map_barchart('B', df_bar, category_col='species', value_col='avg_petal_length',
                     xLabel='Species', yLabel='Average Petal Length',
                     title='Bar Chart: Avg Petal Length by Species')
layout2.display()
```

## 3. Histogram

```python
layout3 = MatrixLayout("H")
layout3.map_histogram('H', df_iris, column='petal_length', bins=20,
                      xLabel='Petal Length', yLabel='Frequency',
                      title='Histogram: Petal Length Distribution')
layout3.display()
```

## 4. Boxplot

```python
layout4 = MatrixLayout("X")
layout4.map_boxplot('X', df_iris, column='petal_width', category_col='species',
                    xLabel='Species', yLabel='Petal Width',
                    title='Boxplot: Petal Width by Species')
layout4.display()
```

## 5. Line Plot

```python
layout5 = MatrixLayout("L")
layout5.map_line('L', df_time, x_col='time_numeric', y_col='value',
                 xLabel='Time', yLabel='Value', title='Line Plot: Time Series')
layout5.display()
```

## 6. Heatmap

```python
corr_matrix = df_iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()

layout6 = MatrixLayout("M")
layout6.map_heatmap('M', corr_matrix, title='Heatmap: Correlation Matrix')
layout6.display()
```

## 7. Pie Chart

```python
df_pie = df_iris['species'].value_counts().reset_index()
df_pie.columns = ['species', 'count']

layout7 = MatrixLayout("P")
layout7.map_pie('P', df_pie, category_col='species', value_col='count',
                title='Pie Chart: Species Distribution')
layout7.display()
```

## 8. Violin Plot

```python
layout8 = MatrixLayout("V")
layout8.map_violin('V', df_iris, value_col='sepal_length', category_col='species',
                   xLabel='Species', yLabel='Sepal Length',
                   title='Violin Plot: Sepal Length by Species')
layout8.display()
```

## 9. Grouped Bar Chart

```python
df_grouped = df_iris.copy()
df_grouped['size'] = pd.cut(df_grouped['petal_length'], bins=2, labels=['Small', 'Large'])

layout9 = MatrixLayout("G")
layout9.map_grouped_barchart('G', df_grouped, main_col='species', sub_col='size',
                             value_col='sepal_width', xLabel='Species', yLabel='Sepal Width',
                             title='Grouped Bar Chart')
layout9.display()
```

## 10. Horizontal Bar Chart

```python
layout10 = MatrixLayout("R")
layout10.map_horizontal_bar('R', df_bar, category_col='species', value_col='avg_petal_length',
                            xLabel='Avg Petal Length', yLabel='Species',
                            title='Horizontal Bar Chart')
layout10.display()
```

## 11. KDE

```python
layout11 = MatrixLayout("K")
layout11.map_kde('K', df_iris, column='petal_length',
                 xLabel='Petal Length', yLabel='Density', title='KDE: Petal Length')
layout11.display()
```

## 12. Distplot

```python
layout12 = MatrixLayout("D")
layout12.map_distplot('D', df_iris, column='sepal_width', bins=20, kde=True, rug=True,
                      xLabel='Sepal Width', yLabel='Density',
                      title='Distplot: Sepal Width')
layout12.display()
```

## 13. Rug Plot

```python
layout13 = MatrixLayout("U")
layout13.map_rug('U', df_iris, column='petal_width', axis='x',
                 xLabel='Petal Width', title='Rug Plot')
layout13.display()
```

## 14. QQ Plot

```python
layout14 = MatrixLayout("Q")
layout14.map_qqplot('Q', df_iris, column='sepal_length', dist='norm',
                    xLabel='Theoretical Quantiles', yLabel='Sample Quantiles',
                    title='QQ Plot')
layout14.display()
```

## 15. ECDF

```python
layout15 = MatrixLayout("E")
layout15.map_ecdf('E', df_iris, column='petal_length',
                  xLabel='Petal Length', yLabel='ECDF', title='ECDF')
layout15.display()
```

## 16. Ridgeline Plot

```python
layout16 = MatrixLayout("I")
layout16.map_ridgeline('I', df_iris, column='petal_length', category_col='species',
                       xLabel='Petal Length', yLabel='Species', title='Ridgeline Plot')
layout16.display()
```

## 17. Ribbon Plot

```python
df_ribbon = df_time.copy()
df_ribbon['upper'] = df_ribbon['value'] + 2
df_ribbon['lower'] = df_ribbon['value'] - 2

layout17 = MatrixLayout("N")
layout17.map_ribbon('N', df_ribbon, x_col='time_numeric', y1_col='lower', y2_col='upper',
                    xLabel='Time', yLabel='Value', title='Ribbon Plot')
layout17.display()
```

## 18. Hexbin Plot

```python
layout18 = MatrixLayout("W")
layout18.map_hexbin('W', df_iris, x_col='sepal_length', y_col='sepal_width',
                    xLabel='Sepal Length', yLabel='Sepal Width', title='Hexbin Plot')
layout18.display()
```

## 19. Error Bars

```python
layout19 = MatrixLayout("Y")
layout19.map_errorbars('Y', df_error, x_col='x', y_col='y', error_col='error',
                       xLabel='X', yLabel='Y', title='Error Bars')
layout19.display()
```

## 20. Step Plot

```python
layout20 = MatrixLayout("T")
layout20.map_step_plot('T', df_time, x_col='time_numeric', y_col='value',
                       xLabel='Time', yLabel='Value', title='Step Plot')
layout20.display()
```

## 21. Parallel Coordinates

```python
layout21 = MatrixLayout("C")
layout21.map_parallel_coordinates('C', df_iris,
    dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    category_col='species', title='Parallel Coordinates')
layout21.display()
```

## 22. Radviz

```python
layout22 = MatrixLayout("Z")
layout22.map_radviz('Z', df_iris,
    features=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    class_col='species', title='Radviz')
layout22.display()
```

## 23. Star Coordinates

```python
layout23 = MatrixLayout("A")
layout23.map_star_coordinates('A', df_iris,
    features=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    class_col='species', title='Star Coordinates')
layout23.display()
```

## 24. 2D Histogram

```python
layout24 = MatrixLayout("F")
layout24.map_hist2d('F', df_iris, x_col='sepal_length', y_col='petal_length',
                    xLabel='Sepal Length', yLabel='Petal Length', title='2D Histogram')
layout24.display()
```

## 25. Polar Plot

```python
df_polar = pd.DataFrame({
    'angle': np.linspace(0, 2*np.pi, 50),
    'radius': np.abs(np.sin(np.linspace(0, 4*np.pi, 50))) * 10
})

layout25 = MatrixLayout("O")
layout25.map_polar('O', df_polar, angle_col='angle', radius_col='radius',
                   title='Polar Plot')
layout25.display()
```

## 26. Funnel Chart

```python
df_funnel = pd.DataFrame({
    'stage': ['Awareness', 'Interest', 'Consideration', 'Intent', 'Purchase'],
    'value': [1000, 800, 600, 400, 200]
})

layout26 = MatrixLayout("J")
layout26.map_funnel('J', df_funnel, stage_col='stage', value_col='value',
                    title='Funnel Chart')
layout26.display()
```

## Test 9: Verificaci√≥n de Comunicaci√≥n JS ‚Üî Python

```python
selection = SelectionModel()

def on_selection_change(change):
    selected_data = change['new']
    print(f"üîî Selecci√≥n: {len(selected_data)} puntos")
    if len(selected_data) > 0:
        print(selected_data[:3])

selection.observe(on_selection_change, names='selected_data')

layout_test = ReactiveMatrixLayout("""
S
B
""", selection_model=selection)

layout_test.set_data(df_iris)

layout_test.add_scatter('S', x_col='petal_length', y_col='petal_width',
                        color_col='species', title='üîç Test Comunicaci√≥n')

layout_test.add_barchart('B', category_col='species', value_col='petal_length',
                         linked_to='S', title='üìä Bar')

layout_test.display()
print("‚úÖ Selecciona puntos para ver mensajes de comunicaci√≥n")
```

## Test 10: Estado del SelectionModel

```python
print("üìä Estado del SelectionModel:")
print(f"Datos seleccionados: {len(selection.selected_data)} puntos")
print(f"√çndices: {selection.selected_indices[:10]}...")

if len(selection.selected_data) > 0:
    selected_df = pd.DataFrame(selection.selected_data)
    print("\nüìà Estad√≠sticas:")
    print(selected_df.describe())
    if 'species' in selected_df.columns:
        print("\nüå∫ Distribuci√≥n por especies:")
        print(selected_df['species'].value_counts())
else:
    print("‚ö†Ô∏è No hay puntos seleccionados")
```
