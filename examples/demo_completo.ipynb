Ejemplos directos y funcionales de MatrixLayout con comunicaci√≥n bidireccional.

```python
import sys, pathlib
root = pathlib.Path().resolve().parents[0]
sys.path.insert(0, str(root))
from BESTLIB import MatrixLayout
from IPython.display import HTML

def show(layout):
    return HTML(layout._repr_html_())

print("‚úÖ BESTLIB listo")
```

```python
# Ejemplo 1: Layout simple
MatrixLayout.set_safe_html(True)
MatrixLayout.map({
    "A": "<b style='color:blue'>Title</b>",
    "C": "<i>Control</i>",
    "B": "<b style='color:red'>ROJO</b>"
})

layout = MatrixLayout("""
AAC
AAC
BBB
""")

show(layout)
```

```python
# Ejemplo 2: Con merge total
MatrixLayout.map({
    "__merge__": True,
    "A": "<b style='color:blue'>Title</b>",
    "C": "<i>Control</i>",
    "B": "<b style='color:red'>ROJO</b>"
})

layout = MatrixLayout("""
AAC
AAC
BBB
""")

show(layout)
```

```python
# Ejemplo 3: Con merge selectivo
MatrixLayout.map({
    "__merge__": ["A", "B"],
    "A": "<b style='color:blue'>Title</b>",
    "C": "<i>Control</i>",
    "B": "<b style='color:red'>ROJO</b>"
})

layout = MatrixLayout("""
AAC
AAC
BBB
""")

show(layout)
```

```python
# Ejemplo 4: Bar chart interactivo
MatrixLayout.map({
    "B": {
        "type": "bar",
        "data": [
            {"category": "Ene", "value": 10},
            {"category": "Feb", "value": 25},
            {"category": "Mar", "value": 18},
            {"category": "Abr", "value": 30}
        ],
        "color": "#4a90e2",
        "interactive": True
    }
})

layout = MatrixLayout("BBB\nBBB")

def on_select(payload):
    print(f"Seleccionado: {payload['items']}")

layout.on('select', on_select)
show(layout)
```

```python
# Ejemplo 5: Scatter plot interactivo
MatrixLayout.map({
    "S": {
        "type": "scatter",
        "data": [
            {"x": 10, "y": 20, "label": "P1", "color": "#e74c3c"},
            {"x": 30, "y": 40, "label": "P2", "color": "#3498db"},
            {"x": 50, "y": 30, "label": "P3", "color": "#2ecc71"},
            {"x": 70, "y": 60, "label": "P4", "color": "#f39c12"},
            {"x": 40, "y": 50, "label": "P5", "color": "#9b59b6"}
        ],
        "pointRadius": 6,
        "interactive": True,
        "zoom": True
    }
})

layout = MatrixLayout("SSS\nSSS")

def on_click(payload):
    print(f"Click: {payload['point']['label']}")

def on_brush(payload):
    print(f"Brush: {len(payload['items'])} puntos")

layout.on('point_click', on_click)
layout.on('select', on_brush)
show(layout)
```

```python
# Ejemplo 6: Bar + Scatter en una celda
MatrixLayout.map({
    "B": {
        "type": "bar",
        "data": [{"category": "A", "value": 10}, {"category": "B", "value": 20}],
        "color": "#4a90e2",
        "interactive": True
    },
    "S": {
        "type": "scatter",
        "data": [{"x": 10, "y": 20, "label": "P1"}, {"x": 30, "y": 40, "label": "P2"}],
        "interactive": True
    }
})

layout = MatrixLayout("BS\nBS")

def handler(payload):
    print(f"Evento: {payload['type']} - {len(payload.get('items', []))} items")

layout.on('select', handler)
layout.on('point_click', handler)
show(layout)
```
