#  BESTLIB - Demo

```python
import sys, pathlib
root = pathlib.Path().resolve().parents[0]
sys.path.insert(0, str(root))
from BESTLIB import MatrixLayout
from IPython.display import HTML

def show(layout):
    return HTML(layout._repr_html_())

print("BESTLIB listo")
```

```python
# Ejemplo 1: Layout simple
MatrixLayout.set_safe_html(True)
MatrixLayout.map({
    "A": "<b style='color:blue'>Title</b>",
    "C": "<i>Control</i>",
    "B": "<b style='color:red'>ROJO</b>"
})

layout = MatrixLayout("""
AAC
AAC
BBB
""")

show(layout)
```

```python
# Ejemplo 2: Con merge total
MatrixLayout.map({
    "__merge__": True,
    "A": "<b style='color:blue'>Title</b>",
    "C": "<i>Control</i>",
    "B": "<b style='color:red'>ROJO</b>"
})

layout = MatrixLayout("""
AAC
AAC
BBB
""")

show(layout)
```

```python
# Ejemplo 3: Con merge selectivo
MatrixLayout.map({
    "__merge__": ["A", "B"],
    "A": "<b style='color:blue'>Title</b>",
    "C": "<i>Control</i>",
    "B": "<b style='color:red'>ROJO</b>"
})

layout = MatrixLayout("""
AAC
AAC
BBB
""")

show(layout)
```
