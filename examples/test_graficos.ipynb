Prueba de bar chart y scatter plot en una sola celda.

```python
import sys, pathlib
root = pathlib.Path().resolve().parents[0]
sys.path.insert(0, str(root))
from BESTLIB import MatrixLayout
from IPython.display import HTML

def show(layout):
    return HTML(layout._repr_html_())

print("âœ… BESTLIB listo para pruebas")
```

```python
# Test: Bar Chart y Scatter Plot juntos
MatrixLayout.map({
    "B": {
        "type": "bar",
        "data": [
            {"category": "A", "value": 100},
            {"category": "B", "value": 150},
            {"category": "C", "value": 120},
            {"category": "D", "value": 180}
        ],
        "color": "#4a90e2",
        "interactive": True
    },
    "S": {
        "type": "scatter",
        "data": [
            {"x": 10, "y": 20, "label": "P1", "color": "#e74c3c"},
            {"x": 25, "y": 35, "label": "P2", "color": "#3498db"},
            {"x": 40, "y": 15, "label": "P3", "color": "#2ecc71"},
            {"x": 30, "y": 45, "label": "P4", "color": "#f39c12"}
        ],
        "pointRadius": 8,
        "interactive": True
    }
})

# Layout lado a lado
test_layout = MatrixLayout("BS")

# Callbacks para capturar eventos
def on_bar_select(payload):
    print(f"ðŸ“Š Bar: {payload['items'][0]['category']} = {payload['items'][0]['value']}")

def on_point_click(payload):
    p = payload['point']
    print(f"ðŸŽ¯ Point: {p['label']} en ({p['x']}, {p['y']})")

test_layout.on('select', on_bar_select)
test_layout.on('point_click', on_point_click)

show(test_layout)
```

```python
# Test: Solo Bar Chart (mÃ¡s grande)
MatrixLayout.map({
    "B": {
        "type": "bar",
        "data": [
            {"category": "Enero", "value": 250},
            {"category": "Febrero", "value": 320},
            {"category": "Marzo", "value": 180},
            {"category": "Abril", "value": 400},
            {"category": "Mayo", "value": 290}
        ],
        "color": "#27ae60",
        "hoverColor": "#229954",
        "interactive": True
    }
})

bar_only = MatrixLayout("B")

def on_bar_click(payload):
    item = payload['items'][0]
    print(f"âœ… Seleccionaste: {item['category']} con valor {item['value']}")

bar_only.on('select', on_bar_click)
show(bar_only)
```

```python
# Test: Solo Scatter Plot (mÃ¡s grande)
MatrixLayout.map({
    "S": {
        "type": "scatter",
        "data": [
            {"x": 5, "y": 10, "label": "Grupo A-1", "color": "#e74c3c"},
            {"x": 8, "y": 15, "label": "Grupo A-2", "color": "#e74c3c"},
            {"x": 12, "y": 8, "label": "Grupo A-3", "color": "#e74c3c"},
            {"x": 25, "y": 30, "label": "Grupo B-1", "color": "#3498db"},
            {"x": 28, "y": 35, "label": "Grupo B-2", "color": "#3498db"},
            {"x": 22, "y": 28, "label": "Grupo B-3", "color": "#3498db"},
            {"x": 45, "y": 20, "label": "Grupo C-1", "color": "#2ecc71"},
            {"x": 48, "y": 25, "label": "Grupo C-2", "color": "#2ecc71"},
            {"x": 42, "y": 18, "label": "Grupo C-3", "color": "#2ecc71"}
        ],
        "pointRadius": 7,
        "interactive": True
    }
})

scatter_only = MatrixLayout("S")

def on_scatter_click(payload):
    p = payload['point']
    print(f"ðŸŽ¯ Clickeaste: {p['label']} en coordenadas ({p['x']}, {p['y']})")

scatter_only.on('point_click', on_scatter_click)
show(scatter_only)
```

```python
# Test: Layout 2x2 con ambos grÃ¡ficos
MatrixLayout.map({
    "B": {
        "type": "bar",
        "data": [
            {"category": "Q1", "value": 85},
            {"category": "Q2", "value": 92},
            {"category": "Q3", "value": 78},
            {"category": "Q4", "value": 96}
        ],
        "color": "#9b59b6",
        "interactive": True
    },
    "S": {
        "type": "scatter",
        "data": [
            {"x": 15, "y": 25, "label": "Alpha", "color": "#e67e22"},
            {"x": 35, "y": 40, "label": "Beta", "color": "#34495e"},
            {"x": 25, "y": 15, "label": "Gamma", "color": "#e74c3c"}
        ],
        "pointRadius": 9,
        "interactive": True
    }
})

layout_2x2 = MatrixLayout("BS\nBS")

def universal_handler(payload):
    if 'items' in payload:  # Bar chart
        print(f"ðŸ“Š Bar event: {len(payload['items'])} items")
    if 'point' in payload:  # Scatter plot
        print(f"ðŸŽ¯ Scatter event: {payload['point']['label']}")

layout_2x2.on('select', universal_handler)
layout_2x2.on('point_click', universal_handler)

show(layout_2x2)
```
