## 📦 Installation & Setup


```python
# Import BESTLIB with debug mode
from BESTLIB.matrix import MatrixLayout
import random
import json

# Enable debug to see communication
MatrixLayout.set_debug(True)

print("✅ BESTLIB Advanced Features Ready!")
```

## 🎯 Feature 1: Automatic Selection Data API

No more manual callbacks! Use `get_selection()` to access data directly.


```python
# Generate sample data
random.seed(42)
bar_data = [
    {"category": f"Category {i}", "value": random.randint(10, 100)}
    for i in range(1, 8)
]

scatter_data = [
    {
        "x": random.randint(0, 100),
        "y": random.randint(0, 100),
        "label": f"Point {i}",
        "category": random.choice(["A", "B", "C"])
    }
    for i in range(20)
]

print(f"📊 Generated {len(bar_data)} bar data points")
print(f"📈 Generated {len(scatter_data)} scatter data points")
```

```python
# Create charts with automatic selection storage
MatrixLayout.map({
    'B': {
        "type": "bar",
        "data": bar_data,
        "color": "#4a90e2",
        "interactive": True,
        "axes": True
    },
    'S': {
        "type": "scatter",
        "data": scatter_data,
        "pointRadius": 5,
        "interactive": True,
        "axes": True
    }
})

# Create layout
layout = MatrixLayout("""
BBBB
SSSS
""")

print("👆 Interact with the charts above, then run the next cell to see selection data!")
layout
```

```python
# Access selection data without callbacks!
bar_selection = layout.get_selection('select')
scatter_selection = layout.get_selection('point_click')

print("🎯 BAR CHART SELECTION:")
if bar_selection:
    print(f"  Items: {len(bar_selection.get('items', []))}")
    print(f"  Data: {bar_selection}")
else:
    print("  No selection yet")

print("\n🎯 SCATTER CHART SELECTION:")
if scatter_selection:
    print(f"  Point: {scatter_selection.get('point', {})}")
    print(f"  Data: {scatter_selection}")
else:
    print("  No point clicked yet")

print("\n💡 Try interacting with the charts above and run this cell again!")
```
